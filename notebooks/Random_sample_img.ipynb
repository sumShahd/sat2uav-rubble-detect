```python
import os
import random
import shutil

# Define the path to your dataset
dataset_path = '/path/to/your/dataset'

# Define the desired number of images to extract
num_images_to_extract = 100

# Define the directory to save the extracted images
output_dir = '/path/to/output/directory'

# Create the output directory if it doesn't exist
os.makedirs(output_dir, exist_ok=True)

# Get a list of all image files in the dataset directory
image_files = [f for f in os.listdir(dataset_path) if os.path.isfile(os.path.join(dataset_path, f))]

# Check if there are enough images in the dataset
if len(image_files) < num_images_to_extract:
    print(f"Warning: The dataset only contains {len(image_files)} images, which is less than the requested {num_images_to_extract}.")
    num_images_to_extract = len(image_files)

# Randomly sample images from the dataset
sampled_images = random.sample(image_files, num_images_to_extract)

# Copy the sampled images to the output directory
for image_name in sampled_images:
    source_path = os.path.join(dataset_path, image_name)
    destination_path = os.path.join(output_dir, image_name)
    shutil.copy2(source_path, destination_path)

print(f"Successfully extracted {num_images_to_extract} images to {output_dir}")
```
