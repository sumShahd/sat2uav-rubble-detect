```python
import os
from PIL import Image
from torchvision import transforms
from tqdm import tqdm
```

```python
input_folder = '/content/drive/MyDrive/uav_eg_data/raw'
output_folder = '/content/drive/MyDrive/uav_eg_data/processed'

# os.makedirs(output_folder, exist_ok=True)
```

```python
transform = transforms.Compose([
    transforms.Resize(512),  # resize smaller edge to 512
    transforms.CenterCrop(512),  # crop to 512×512 (keeps central content)
    transforms.ToTensor()
])
```

```python
for filename in tqdm(os.listdir(input_folder)):
    if filename.lower().endswith(('.png', '.jpg', '.jpeg', '.tiff', '.bmp')):
        img_path = os.path.join(input_folder, filename)
        img = Image.open(img_path).convert("RGB")
        tensor_img = transform(img)

        # Convert back to PIL and save
        save_img = transforms.ToPILImage()(tensor_img)
        save_img.save(os.path.join(output_folder, filename))
```
