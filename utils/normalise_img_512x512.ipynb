```python
import os
from PIL import Image
from torchvision import transforms
from tqdm import tqdm
```

```python
inputFolder = '/content/drive/MyDrive/uav_eg_data/raw'
outputFolder = '/content/drive/MyDrive/uav_eg_data/processed'

# os.makedirs(outputFolder, exist_ok=True)
```

```python
transform = transforms.Compose([
    transforms.Resize(512),  # resizing smaller edge to 512
    transforms.CenterCrop(512),  # cropping to 512×512 (keeps stuff in the centre)
    transforms.ToTensor()
])
```

```python
for filename in tqdm(os.listdir(inputFolder)):
    if filename.lower().endswith(('.png', '.jpg', '.jpeg', '.tiff', '.bmp')):
        imgPath = os.path.join(inputFolder, filename)
        img = Image.open(imgPath).convert("RGB")
        tensorImg = transform(img)

        # Convert back to PIL and save
        finalImg = transforms.ToPILImage()(tensorImg)
        finalImg.save(os.path.join(outputFolder, filename))
```
