#GITHUB DATA

```python
import requests
import zipfile
import io
import os
import shutil

githubRepo = "https://github.com/Marwen200/Gaza_deep_dav/" # Replace with the actual GitHub repo URL
zipFilename = "no_damage_part2.zip" # Replace with the actual zip file name in the repo
driveFolderPath = '/content/drive/MyDrive/training_data/GAZADeepDaV/no_damage_2' # Destination folder in Google Drive

# Construct the raw file URL
rawUrl = githubRepo.replace("github.com", "raw.githubusercontent.com") + "/main/" + zipFilename # Assuming the file is in the main branch

print(f"Downloading {zipFilename} from {rawUrl}")

if not os.path.exists(driveFolderPath):
    os.makedirs(driveFolderPath)
    print(f"Created destination folder: {driveFolderPath}")

try:
    response = requests.get(rawUrl)
    response.raise_for_status()  # Raise an exception for bad status codes

    # Step 3: Unzip files and Step 4/5: Identify and Copy image files directly to Drive
    imageExtensions = ['.jpg', '.jpeg', '.png', '.gif', '.bmp']
    copyCount = 0

    with zipfile.ZipFile(io.BytesIO(response.content)) as zip_ref:
        for fileInfo in zip_ref.infolist():
            if not fileInfo.is_dir() and any(fileInfo.filename.lower().endswith(ext) for ext in imageExtensions):
                # Construct the destination path, preserving the original filename
                dataDstPath = os.path.join(driveFolderPath, os.path.basename(fileInfo.filename))
                try:
                    # Extract the file directly to the drive folder
                    with zip_ref.open(fileInfo) as source, open(dataDstPath, "wb") as target:
                        shutil.copyfileobj(source, target)
                    copyCount += 1
                except Exception as e:
                    print(f"Error copying file {fileInfo.filename} to {dataDstPath}: {e}")

    print(f"Finished copying {copyCount} image files to {driveFolderPath}")

except requests.exceptions.RequestException as e:
    print(f"Error downloading the file: {e}")
except zipfile.BadZipFile:
    print("Error: The downloaded file is not a valid zip file.")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
```
